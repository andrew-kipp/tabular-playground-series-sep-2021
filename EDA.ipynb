```python
# Describe the data to determine if features are already scaled or if they need to be scaled/normalized
# 'id' and 'claim' are integer values, all others are float; id will not be used as a predictor and claim is the binary target
# Feature columns have a wide range of values and should be scaled/normalized; scaling/normalizing the features will 
# prevent a model from being too depending on features with much larger values
training_data.describe()
```

```python
# Check for any features which might possibly be categorical, the selection of "< 0.01 * number of unique ids" was just to limit output 
unique_ids = len(training_data['id'].unique())
for col in training_data.columns:
    unique_vals = len(training_data[col].unique())
    if unique_vals < 0.01 * unique_ids:
        print('Column: {}, Unique Values: {}'.format(col, unique_vals))
```

```python
# Check the target variable to determine if the classes are balanced - they are
training_data.claim.value_counts(normalize = True)
```

```python
# Check that the claim values aren't directly related to the ID value 
# or that there is a split in the data such that target values are not 
# randomly distributed among the IDs
# The Id value will not be used for prediction but it is good to know 
# if the claim value is related because it may make the modeling 
# process moot
# The average and median ID values per target group suggests that the
# IDs and claims values are unrelated. Visual inspection of the data
# also shows no obvious pattern between the IDs and claims
print('Average ID value for claim groups: \n', training_data.groupby(['claim']).id.mean())
print('Median ID value for claim groups: \n', training_data.groupby(['claim']).id.median())
```

```python
# Check missing values to determine prevalence and if imputation is necessary - it is
print(training_data.isnull().sum())
print('Total Nulls:', training_data.isnull().sum().sum())
print('Total Values:', training_data.isnull().count().sum())
print('Fraction Nulls:', training_data.isnull().sum().sum() / training_data.isnull().count().sum())
```
